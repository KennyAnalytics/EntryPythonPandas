```python
# Importing the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading the csv file
df = pd.read_csv('mega.csv')
```

```python
# Viewing the first 5 columns
df.head()
```

```python
# No of rows and columns
df.shape
```

```python
df.columns
```

```python
df.info()
```

```python
mega = df.copy()  
# create a copy of the original DataFrame
mega['Type 2'].fillna(df['Type 1'], inplace=True)  
# fill missing values in 'Type 2' with values from 'Type 1'
```

```python
df.head()
```

```python
#we will be using mega going forward
mega.head()
```

```python
# we see that there is no null in 'Type 2'
mega.info()
```

```python
# Analyze the distribution of Mega Pokemon Types
type_count = mega['Type 1'].value_counts()
print(type_count)
```

```python
# Analyze the distribution of Pokemon stats
stat_cols = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
stat_df = mega[stat_cols]
print(stat_df.describe().round(2))
```

```python
# Explore the relationship between Pokemon types and stats
type_stat_df = mega.groupby('Type 1')[stat_cols].mean()
print((type_stat_df).round(2))
```

```python
# Generation is the release date of certain pokemon and it is more front loaded towards Generation 1 and Generation 3 due to the timing of certain game release dates
mega['Generation'].value_counts().sort_index(ascending=True)
```

```python
legendary_counts = mega['Legendary'].value_counts()
legendary_counts
```

```python
plt.bar(x=legendary_counts.index.astype(str), height=legendary_counts.values)
plt.xlabel('Legendary')
plt.ylabel('Count')
plt.title('Legendary Counts')
for i, v in enumerate(legendary_counts.values):
    plt.text(i, v, str(v), ha='center', va='bottom')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.boxplot(stat_df.values, labels=stat_cols)
plt.title('Distribution of Pokemon Stats')
plt.xlabel('Stat')
plt.ylabel('Value')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.bar(type_count.index, type_count.values)
plt.title('Distribution of Pokemon Types')
plt.xlabel('Pokemon Type')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.bar(type_stat_df.index, type_stat_df['Attack'], label='Attack')
plt.bar(type_stat_df.index, type_stat_df['Defense'], label='Defense')
plt.legend()
plt.title('Average Stats by Pokemon Type')
plt.xlabel('Pokemon Type')
plt.ylabel('Value')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(type_stat_df.index, type_stat_df['Attack'], label='Attack')
plt.scatter(type_stat_df.index, type_stat_df['Defense'], label='Defense')
plt.legend()
plt.title('Average Stats by Pokemon Type')
plt.xlabel('Pokemon Type')
plt.ylabel('Value')
plt.show()
```
